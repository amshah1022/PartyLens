```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your cleaned dataset
df = pd.read_csv("../data/party_data_cleaned.csv")

# Ensure date is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Quick preview
df.head()
```

```python
sns.histplot(df['instagram_posts'], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Instagram Posts")
plt.xlabel("Instagram Posts")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='turnout_level', data=df, palette='pastel')
plt.title("Event Turnout Level Distribution")
plt.xlabel("Turnout Level")
plt.ylabel("Number of Events")
plt.show()
```

```python
sns.boxplot(x='turnout_level', y='avg_likes', data=df, palette='coolwarm')
plt.title("Average Likes by Turnout Level")
plt.xlabel("Turnout Level")
plt.ylabel("Average Likes")
plt.show()
```

```python
sns.boxplot(x='event_time', y='instagram_posts', data=df, palette='viridis')
plt.title("Instagram Posts by Event Time")
plt.xlabel("Event Time")
plt.ylabel("Instagram Posts")
plt.show()
```

```python
sns.countplot(x='weather_condition', hue='turnout_level', data=df, palette='Set2')
plt.title("Weather Condition vs Turnout Level")
plt.xlabel("Weather")
plt.ylabel("Number of Events")
plt.legend(title='Turnout Level')
plt.show()
```
